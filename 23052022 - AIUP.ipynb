```python
import numpy as np
from scipy import linalg

A = np.array([[1,2],[3,4]])
print(A)
B = linalg.inv(A)
print(B)
```

```python
#Denklem takımlarının çözümü
A = np.array([[1,3,5],[2,5,1],[2,3,8]])
print(A)
b = np.array([[10], [8], [3]])
print(b)
print("1 - scipy çözümü :")
print(linalg.inv(A).dot(b)) #Yavaş çözüm
print("2 - numpy solve çözümü :")
print(np.linalg.solve(A, b)) #Hızlı çözüm
```

```python
#İntegral
from scipy import integrate

x2 = lambda x : x**2
I = integrate.quad(x2, 0, 6)
print(I)
```

```python
import scipy.integrate as integrate
import scipy.special as special
result = integrate.quad(lambda x : special.jv(2.5, x), 0, 4.5)
print(result)
```

```python
from scipy.misc import derivative

def f(x):
    return x**3 + x**2

D = derivative(f, 1.0, dx = 1e-6)
print(D)
```

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.tan(x)
y3 = np.exp(x)
y4 = np.power(x, 3)

fig1, ax1 = plt.subplots(nrows = 2, ncols = 2, figsize = (12,12))
ax1[0, 0].plot(x, y1, "tab:blue")
ax1[0, 0].set_title("Square Function")
ax1[0, 0].set_xlabel(r'$X$', fontsize = 18)
ax1[0, 0].set_ylabel(r'$Y$', fontsize = 18)

ax1[0, 1].plot(x, y2, "r-")
ax1[0, 1].set_title("Tangent Function")
ax1[0, 1].set_xlabel(r'$X$', fontsize = 18)
ax1[0, 1].set_ylabel(r'$Y$', fontsize = 18)

ax1[1, 0].plot(x, y3, "k-")
ax1[1, 0].set_title("Exponential Function")
ax1[1, 0].set_xlabel(r'$X$', fontsize = 18)
ax1[1, 0].set_ylabel(r'$Y$', fontsize = 18)

ax1[1, 1].plot(x, y4, "tab:orange")
ax1[1, 1].set_title("Cubic Function")
ax1[1, 1].set_xlabel(r'$X$', fontsize = 18)
ax1[1, 1].set_ylabel(r'$Y$', fontsize = 18)

plt.show()
```
